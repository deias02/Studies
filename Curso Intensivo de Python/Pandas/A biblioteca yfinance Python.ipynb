# A biblioteca yfinance Python

O `yfinance` é uma biblioteca Python com uma interface amigável para download de dados históricos de mercado do Yahoo Finance. Ele permite obter preços históricos de ações, dividendos e outros dados financeiros de ações, fundos negociados em bolsa (ETFs) e outros títulos.

Este exemplo mostra o código para usar yfinance para baixar preços históricos de ações.

```python
import yfinance as yf

# Download historical data for a stock
msft = yf.Ticker("MSFT")

msft_data = msft.history(period="max")

# Display the downloaded data
msft_data.head()
```

**Explicação para o código acima:**

Primeiro, importe a biblioteca `yfinance` usando o alias `yf`.
Em seguida, crie um objeto Ticker para as ações da Microsoft `(“MSFT”)`.
Use o método `history` do objeto `Ticker` para baixar os dados históricos da ação. O parâmetro `períod` do método `history` especifica quando você deseja fazer download dos dados. Neste exemplo, está definido como máximo para baixar o `max` de dados históricos disponíveis.

Aqui estão alguns dos valores possíveis para o parâmetro de período e o que eles representam:

* period="1d": Baixe 1 dia de dados históricos.
* period="5d": Baixe 5 dias de dados históricos.
* period="1mo": Baixe 1 mês de dados históricos.
* period="3mo": Baixe 3 meses de dados históricos.
* period="6mo": Baixe 6 meses de dados históricos.
* period="1y": Baixe 1 ano de dados históricos.
* period="2y": Baixe 2 anos de dados históricos.
* period="5y": Baixe 5 anos de dados históricos.
* period="10y": Baixe 10 anos de dados históricos.
* period="acumulado": Baixe dados históricos desde o início do ano atual.
* period="max": Baixe todos os dados históricos disponíveis.

Finalmente, você imprime os dados baixados usando a função `head`. Esses dados baixados exibirão um quadro de dados do Pandas contendo os preços históricos das ações da Microsoft e outros dados financeiros.

```python
msft_data = msft.history(period="10y")
msft_data.head()
```
