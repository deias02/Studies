As funções **"Metódos Get"** mencionadas no início deste notebook podem gerar os parâmetros necessários para construir um conjunto de dados. Por exemplo, para obter o nome completo, chamamos a função `get_full_name()`.

```python
name = r.get_full_name()
```

```python
name
```

Digamos que precisamos apenas de 10 usuários com nomes completos e endereços de e-mail. Podemos escrever um "for-loop" para imprimir esses 10 usuários.

```python
for user in some_list:
    print (user.get_full_name()," ",user.get_email())
```

Para gerar uma tabela com informações sobre os usuários, podemos escrever uma função contendo todos os parâmetros desejáveis. Por exemplo, nome, sexo, cidade, etc. Os parâmetros dependerão dos requisitos do teste a ser realizado. Chamamos os Métodos Get, listados no início deste caderno. Em seguida, retornamos o dataframe do pandas com os usuários.

```python
def get_users():
    users =[]
     
    for user in RandomUser.generate_users(10):
        users.append({"Name":user.get_full_name(),"Gender":user.get_gender(),"City":user.get_city(),"State":user.get_state(),"Email":user.get_email(), "DOB":user.get_dob(),"Picture":user.get_picture()})
      
    return pd.DataFrame(users)     
```

## Exemplo 2: API Fruityvice

Outra forma mais comum de usar APIs é por meio da biblioteca `requests`. O próximo laboratório, Solicitações e HTTP, conterá mais informações sobre solicitações.

Começaremos importando todas as bibliotecas necessárias.

```python
import requests
import json
```

Obteremos os dados da API [fruityvice](https://www.fruityvice.com) usando a função `requests.get("url")`. Os dados estão em formato json.

```python
data = requests.get("https://fruityvice.com/api/fruit/all")
```

Recuperaremos os resultados usando a função `json.loads()`.

```python
results = json.loads(data.text)
```

Converteremos seus dados json em dataframe do pandas

```python
pd.DataFrame(results)
```

The result is in a nested json format. The 'nutrition' column contains multiple subcolumns, so the data needs to be 'flattened' or normalized.

```python
df2 = pd.json_normalize(results)
```
